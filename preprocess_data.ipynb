# Data cleaning and preprocessing

```python
import pandas as pd

data = pd.read_excel('data/online_retail.xlsx')
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
data = data[data['Price'] > 0]
data
```

```python
data = data[data['Quantity'] > 0]
data
```

```python
data = data.drop(['StockCode', 'Description', 'Country'], axis=1)
data.columns
```

```python
data = data.drop_duplicates()
data
```

```python
data = data[data['Customer ID'].notna()]
data
```

# RFM implementation

```python
amount = pd.DataFrame(data.Quantity * data.Price, columns = ['Amount'])

amount
```

```python
data = pd.concat([data, amount], axis=1)
data
```

```python
monetary = data.groupby('Customer ID').Amount.sum()
monetary = monetary.reset_index()
monetary
```

```python
frequency = data.groupby('Customer ID').Invoice.count()
frequency = frequency.reset_index()
frequency
```

```python
recency = (data.groupby('Customer ID').InvoiceDate.max() - data.InvoiceDate.min()).reset_index()
recency.columns = ['Customer ID', 'Recency']
recency.Recency = recency.Recency.dt.days
recency
```

```python
rfm = pd.merge(recency, frequency, on='Customer ID')
rfm = pd.merge(rfm, monetary, on='Customer ID')
rfm.columns = ['Customer ID', 'Recency', 'Frequency', 'Monetary']
rfm
```

# Outlier detection

```python
import matplotlib.pyplot as plt

def plot_rfm(rfm):
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    ax[0].hist(rfm.Recency, bins=20)
    ax[0].set_title('Recency')
    ax[1].hist(rfm.Frequency, bins=20)
    ax[1].set_title('Frequency')
    ax[2].hist(rfm.Monetary, bins=20)
    ax[2].set_title('Monetary')
    plt.show()

plot_rfm(rfm)
```

```python
def delete_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    df = df[(df[col] >= q1 - 1.5 * iqr) & (df[col] <= q3 + 1.5 * iqr)]
    return df

rfm = delete_outliers(rfm, 'Monetary')
rfm = delete_outliers(rfm, 'Frequency')
rfm = delete_outliers(rfm, 'Recency')

rfm
```

```python
plot_rfm(rfm)
```

# Standardization and normalization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm[['Recency', 'Frequency', 'Monetary']] = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
rfm
```
